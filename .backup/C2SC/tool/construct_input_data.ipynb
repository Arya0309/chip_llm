````python
datalist_dir = "/home/S113062628/project/chip_llm/chip_llm/systemc_generation/dataset/geek/list.csv"
dataset_dir = (
    "/home/S113062628/project/chip_llm/chip_llm/systemc_generation/dataset/geek/data"
)
testbench_dir = "/home/S113062628/project/chip_llm/chip_llm/systemc_generation/dataset/geek/testbench"
output_file = "/home/S113062628/project/chip_llm/chip_llm/systemc_generation/input_data/qwen/zero_shot.json"

prompt_dir = "/home/S113062628/project/chip_llm/chip_llm/systemc_generation/prompts/self_debug.json"
````

````python
import json
import os
import pandas as pd
import re

df_input = pd.read_csv(datalist_dir)

# Insert the code into the DataFrame
list_code = []

for code in os.listdir(dataset_dir):
    if code.endswith(".cpp"):
        with open(os.path.join(dataset_dir, code), "r") as f:
            code_content = f.read()
        list_code.append({"task": code.split(".")[0], "code": code_content})

df_code = pd.DataFrame(list_code)

df_input = df_input.merge(df_code, on="task", how="left")

# Insert signal, module_name and testbench into the DataFrame
list_testbench_info = []
for dir in os.listdir(testbench_dir):
    for file in os.listdir(os.path.join(testbench_dir, dir)):
        if file.endswith(".cpp"):
            with open(os.path.join(testbench_dir, dir, file), "r") as f:
                testbench_content = f.read()

                module_match = re.search(
                    r"\bSC_MODULE\s*\(\s*([A-Za-z_]\w*)\s*\)", testbench_content
                )
                module_name = module_match.group(1) if module_match else None

                lines = testbench_content.splitlines()
                signals = []
                for line in lines:
                    if "sc_in" in line or "sc_out" in line:
                        clean_line = re.sub(r"//.*", "", line).strip()
                        if clean_line:
                            signals.append(clean_line)
                signal = "\n".join(signals)
                if not signal:
                    raise ValueError(
                        f"No signals found in testbench file: {os.path.join(testbench_dir, dir, file)}"
                    )

                testbench = (
                    "SC_MODULE(Testbench)"
                    + testbench_content.split("SC_MODULE(Testbench)")[1]
                )

                list_testbench_info.append(
                    {
                        "task": dir,
                        "module_name": module_name,
                        "testbench": testbench,
                        "signal": signal,
                        "adoption": True,
                        "reason": None,
                    }
                )
df_input = df_input.merge(pd.DataFrame(list_testbench_info), on="task", how="left")

df_input["adoption"] = df_input["adoption"].fillna(False).astype(bool)
df_input["task_id"] = df_input.index + 1

df_input
````

````python
BASIC = """#include <systemc.h>
// ...
// <systemC code here>
// ..."""

MATH = """#include <systemc.h>

// Module that adds two integers
SC_MODULE(Adder) {
    sc_in<int> a;  // Input port for the first number
    sc_in<int> b;  // Input port for the second number
    sc_out<int> sum; // Output port for the sum

    // Constructor
    SC_CTOR(Adder) {
        // Process to perform addition
        SC_METHOD(add);
        dont_initialize();
        sensitive << a << b;
    }

    // Method to add the two numbers
    void add() {
        sum.write(a.read() + b.read());
    }
};"""

ARRAY = """#include <systemc.h>

// Module: BubbleSort
// Sorts a small fixed-size array of integers using the bubble sort algorithm.
SC_MODULE(BubbleSort) {
    sc_in<bool>  clk;      // Clock signal
    sc_in<bool>  start;    // Start signal to begin sorting
    sc_out<bool> done;     // Signal asserted when sorting is complete

    sc_in<int>   in_arr[4];  // Input array of 4 integers to be sorted
    sc_out<int>  out_arr[4]; // Output array of 4 sorted integers

    int buf[4];  // Internal buffer to hold array elements during sorting

    // Constructor: registers the sorting process with the clock
    SC_CTOR(BubbleSort) {
        SC_THREAD(sort_proc);
        sensitive << clk.pos();
    }

    // Thread: waits for start, performs bubble sort, writes results, toggles done
    void sort_proc() {
        const int N = 4;
        while (true) {
            wait();  // wait for next clock edge
            if (start.read()) {
                // Read inputs into internal buffer
                for (int i = 0; i < N; ++i)
                    buf[i] = in_arr[i].read();

                // Perform bubble sort on buffer[]
                for (int i = 0; i < N - 1; ++i)
                    for (int j = 0; j < N - i - 1; ++j)
                        if (buf[j] > buf[j + 1])
                            std::swap(buf[j], buf[j + 1]);

                // Write sorted results to output ports
                for (int i = 0; i < N; ++i)
                    out_arr[i].write(buf[i]);

                // Signal done for one cycle
                done.write(true);
                wait();
                done.write(false);
            }
        }
    }
};"""

MATRIX = """#include <systemc.h>

// Module: MatrixAdder
// Adds two 4×4 matrices of integers element-wise.
SC_MODULE(MatrixAdder) {
    sc_in<int> A[4*4];   // Flattened 4×4 input matrix A (16 elements)
    sc_in<int> B[4*4];   // Flattened 4×4 input matrix B (16 elements)
    sc_out<int> C[4*4];  // Flattened 4×4 output matrix C = A + B

    // Constructor: registers a combinational add method sensitive to all inputs
    SC_CTOR(MatrixAdder) {
        SC_METHOD(do_add);
        // Make the add method sensitive to every element of A and B
        for (int i = 0; i < 4*4; i++) {
            sensitive << A[i] << B[i];
        }
    }

    // Method: reads A[i] and B[i], writes sum into C[i] for all 16 positions
    void do_add() {
        for (int i = 0; i < 4*4; i++) {
            C[i].write(A[i].read() + B[i].read());
        }
    }
};"""

ALL = (
    "Few_shot_example_1:\n"
    + MATH
    + "\n\n"
    + "Few_shot_example_2:\n"
    + ARRAY
    + "\n\n"
    + "Few_shot_example_3:\n"
    + MATRIX
    + "\n\n"
)
````
